```python
from lumos.model.ptm_new.core import *
io_cmos = IOCore(tech=22)
io_tfet = IOCore_TFET(tech=22)
print 'Area: ', io_cmos.area, io_tfet.area
print 'Perf(freq): ', io_cmos.perf, io_tfet.perf
print 'Vdd(mV): ', io_cmos.vnom, io_tfet.vnom
print 'Power: ', io_cmos.power, io_tfet.power
cmos_perf = []
for vdd in xrange(io_cmos.vmin, io_cmos.vmax):
    io_cmos.vdd = vdd
    cmos_perf.append(io_cmos.perf)
tfet_perf = []
for vdd in xrange(io_tfet.vmin, io_tfet.vmax):
    io_tfet.vdd = vdd
    tfet_perf.append(io_tfet.perf)
plt.plot(range(io_cmos.vmin, io_cmos.vmax), cmos_perf, label='CMOS')
plt.plot(range(io_tfet.vmin, io_tfet.vmax), tfet_perf, label='TFET')
plt.legend(loc='best')
plt.yscale('log')
    
```
